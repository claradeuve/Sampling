## Exercici 1
Agafa un conjunt de dades de tema esportiu que t'agradi. Realitza un mostreig de les dades generant una mostra aleatòria simple i una mostra sistemàtica

```python
import pandas as pd
import numpy as np

df_birds = pd.read_csv("Minnesota Birds.csv")
df_birds.info()
sample_size = int(len(df_birds)*0.2)
```

**Mostra aleatòria simple**

```python
df_birds_simple = df_birds.sample(n=sample_size, replace=True)
display(df_birds_simple)
```

**Mostra sistemàtica**

```python
import random

def systematic_sampling(df, step):
    start = random.randint(0, len(df))
    indexes = np.arange(start, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

df_birds_syst = systematic_sampling(df_birds, len(df_birds)/sample_size)
display(df_birds_syst)
```

## Exercici 2
Continua amb el conjunt de dades de tema esportiu i genera una mostra estratificada i una mostra utilitzant SMOTE (Synthetic Minority Oversampling Technique)

**Mostra estratificada**

```python
df_birds_strat = df_birds.groupby('Family', group_keys=False).apply(lambda x: x.sample(frac=0.2))
display(df_birds_strat)
```

**SMOTE (Synthetic Minority Oversampling Technique)**

*Con reducción de muestra*

```python
import matplotlib.pyplot as plt

df_birds.info()
df_birds.State.value_counts().plot(kind='bar', title='Count (target)')
plt.show()
```

```python
df_birds['State'] = df_birds['State'].fillna('undefined')
df_birds.info()

df_birds.State.value_counts().plot(kind='bar', title='Count (target)')
plt.show()
```

```python
count_class_undefined, count_class_minnesota = df_birds.State.value_counts()

# Divide by class
df_class_undefined = df_birds[df_birds['State'] == 'undefined'] #majority class
df_class_minnesota = df_birds[df_birds['State'] == 'Minnesota'] #minority class

# Sample Majority class (y=0, to have same number of records as minority calls (y=1)
df_class_undefined_red = df_class_undefined.sample(count_class_minnesota)
# join the dataframes containing y=1 and y=0
df_test_red = pd.concat([df_class_undefined_red, df_class_minnesota])

print("Reducción aleatoria de la muestra:")
print(df_test_red.State.value_counts(), "\n")
print("Cantidad de registros:", df_test_red.shape[0])

df_test_red.State.value_counts().plot(kind='bar', title='Count (target)');
```

*Con incremento de muestra*

```python
df_birds.State.value_counts().plot(kind='bar', title='Count (target)')
plt.show()

df_class_minnesota_over = df_class_minnesota.sample(count_class_undefined, replace=True)
df_test_incr = pd.concat([df_class_undefined, df_class_minnesota_over], axis=0)

print("Incremento aleatorio de la muestra:")
print(df_test_incr.State.value_counts(), "\n")
print("Cantidad de registros:", df_test_incr.shape[0])

df_test_incr.State.value_counts().plot(kind='bar', title='Count (target)');
```
